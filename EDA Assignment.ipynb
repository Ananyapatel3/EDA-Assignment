```python
Q1: Dimensions of the Flight Price Dataset
#The dataset typically contains rows representing individual flights and columns representing flight attributes such as airline, departure time, stops, duration, and price. You can check the dimensions using:
import pandas as pd

df = pd.read_csv("flight_price_dataset.csv")
print(df.shape)  # Outputs (num_rows, num_columns)
```

```python
Q2: Distribution of Flight Prices
#To visualize the distribution of flight prices, you can create a histogram
import matplotlib.pyplot as plt

plt.hist(df["Price"], bins=30, edgecolor="black")
plt.xlabel("Flight Price")
plt.ylabel("Frequency")
plt.title("Distribution of Flight Prices")
plt.show()

Q3: Range of Flight Prices
To find the minimum and maximum price:
print(f"Minimum Price: {df['Price'].min()}")
print(f"Maximum Price: {df['Price'].max()}")


Q4: Price Variation by Airline
A boxplot helps compare flight prices across different airlines:
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x=df["Airline"], y=df["Price"])
plt.xticks(rotation=45)
plt.xlabel("Airline")
plt.ylabel("Flight Price")
plt.title("Flight Price Variation by Airline")
plt.show()


Q5: Identifying Outliers
Outliers can be detected using a boxplot:
sns.boxplot(y=df["Price"])
plt.title("Boxplot of Flight Prices")
plt.show()


Outliers may indicate special promotions, last-minute bookings, or unusually expensive flights. They can skew analysis, so handling them carefully is crucial.
Q6: Identifying Peak Travel Season
To determine peak travel seasons, analyze:
- Month and day of travel
- Demand fluctuations
- Holiday periods
- Weather conditions affecting travel
A time-series plot can help visualize seasonal trends.
Q7: Identifying Flight Price Trends
Key features to analyze:
- Booking time vs. departure time
- Seasonal variations
- Airline pricing strategies
- Demand fluctuations
Visualizations like line plots, heatmaps, and moving averages can reveal trends.
Q8: Factors Affecting Flight Prices
Important factors include:
- Airline
- Number of stops
- Departure time
- Booking window
- Seasonality
- Demand trends
```
